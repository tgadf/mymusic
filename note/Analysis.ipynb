# Master Music Data

```python
from pandas import DataFrame, Series
def getMusicDF(mmb):
    musicDict = {}
    for artistName in mmb.getArtists():
        idx = mdbmaps["Music"].getPrimaryKey(artistName=artistName)
        mmbArtistAlbums = mmb.getArtistAlbumsByArtist(artistName)
        numArtistAlbums = {key: val.getNumAlbums() for key,val in mmbArtistAlbums.items()}
        numSongs        = numArtistAlbums["Random"]
        numAlbums       = sum(numArtistAlbums.values()) - numSongs    
        musicDict[artistName] = {"Songs": numSongs, "Albums": numAlbums, "Full": numArtistAlbums}
    return DataFrame(musicDict).T

def getMusicDBDF(mdbmap):
    musicDict = {}
    for idx,artistData in mdbmap.get().items():
        artistName = artistData.getArtistName()
        dbData     = artistData.getDict()
        musicDict[artistName] = dbData
    return DataFrame(musicDict).T

def createMusicDF(mmb, mdbmap):
    df1 = getMusicDF(mmb)
    print("Found Music DF:  {0}".format(df1.shape[0]))
    df2 = getMusicDBDF(mdbmap)
    print("Music Map DB DF: {0}".format(df2.shape[0]))
    musicDF = df1.join(df2).copy(deep=True)
    print("Music Map DF:    {0}".format(musicDF.shape[0]))


    tmp = musicDF[["Songs", "Albums", "Full"]]
    retval = {}
    for artistName,vals in tmp.iterrows():
        key = (artistName,1)
        retval[key] = vals
    musicDF = DataFrame(retval).T
    musicDF.columns = ["My Songs", "My Albums", "My Full"]
    print("Created MusicDF With {0} Rows".format(musicDF.shape[0]))
    return musicDF
```

```python
def addMusicInfo(dbMatrix, mdf, mType):
    keyIdx = {}
    for idx, row in mdf.iterrows():
        artistName = row["ArtistName"]
        key = (artistName,1)
        while keyIdx.get(key) is not None:
            key = (artistName,key[1]+1)
        keyIdx[key] = True
        if dbMatrix.get(key) is None:
            dbMatrix[key] = {}
        dbMatrix[key][mType] = {}

        dbVals = {}
        for db,dbID in row.iteritems():
            if db in ["DBMatches", "ArtistName"]:
                continue
            dbVals[db] = (dbID,idx)
        dbMatrix[key][mType] = dbVals
        
        
dbMatrix = {}
addMusicInfo(dbMatrix,  mdfs["Music"], "Music")
if mdfs.get("Top40") is not None:
    addMusicInfo(dbMatrix,  mdfs["Top40"], "Top40")
if mdfs.get("Billboard") is not None:
    addMusicInfo(dbMatrix,  mdfs["Billboard"], "Billboard")
if mdfs.get("BillboardYE") is not None:
    addMusicInfo(dbMatrix,  mdfs["BillboardYE"], "BillboardYE")
if mdfs.get("Spotify") is not None:
    addMusicInfo(dbMatrix,  mdfs["Spotify"], "Spotify")
ddbM = DataFrame(dbMatrix).T
ddbM.head()
```

```python
def splitAlbums(x):
    retval = {}
    for db in x.keys():
        if isinstance(x[db],dict):
            retval[db] = dbaat.splitArtistAlbums(db,x[db])
        else:
            retval[db] = {}
    return retval


def mergeSplitAlbums(x, **kwargs):
    retval    = splitAlbums(x)
    dbsToKeep = kwargs["dbsToKeep"]
    
    dbretval = {i: [] for i in range(1,5)}
    for db,dbAlbums in retval.items():
        if dbsToKeep is not None:
            if db not in dbsToKeep:
                continue
        for albumType,albumTypeData in dbAlbums.items():
            dbretval[albumType] += albumTypeData

    dbretval = {albumType: list(set(albumTypeData)) for albumType,albumTypeData in dbretval.items()}
    return dbretval
```

```python
from collections import Counter

def genre(x):
    genres = Counter()
    for albumType,albumTypeData in x.iteritems():
        for album in albumTypeData:
            types = []
            if "Orchestra" in album:
                types.append("Classical")
            if "Blues" in album:
                types.append("Blues")
            if "Jazz" in album:
                types.append("Jazz")
            if sum(x in album for x in ["Music From The ", "Soundtrack", "Theme From"]) > 0:
                types.append("Soundtrack")
                
            for t in types:
                genres[t] += 1

    return genres.most_common(1)
genres = masterArtistSplitAlbumsDF.apply(genre, axis=1)
```

```python
def getNumberOfAlbums(x):
    retval = {albumType: len(albumTypeData) for albumType,albumTypeData in x.to_dict().items() if albumType}
    return retval
masterArtistNumAlbumsDF = DataFrame(masterArtistSplitAlbumsDF.apply(getNumberOfAlbums , axis=1))
masterArtistNumAlbumsDF = masterArtistNumAlbumsDF[0].apply(pd.Series).copy(deep=True)
masterArtistNumAlbumsDF["Genre"] = genres
```

```python
from listUtils import getFlatList
from pandas import DataFrame, Series


def getArtistAlbums(mmb, artistName):
    artistAlbumsDict = mmb.getArtistAlbumsByArtist(artistName)
    artistAlbumsByType = {}
    for albumType,albumsInfo in artistAlbumsDict.items():
        artistAlbumsByType[albumType] = getFlatList(albumsInfo.getAlbums().values())
        
    artistAlbums = []
    albumTypes = ["BoxSet", "Bootleg", "Match", "UnMatched"]
    for albumType in albumTypes:
        if isinstance(artistAlbumsByType.get(albumType), list):
            artistAlbums += [artistAlbumsByType[albumType]]
    return getFlatList(artistAlbums)


def getArtistSongs(mmb, artistName):
    artistAlbumsDict = mmb.getArtistAlbumsByArtist(artistName)
    artistAlbumsByType = {}
    for albumType,albumsInfo in artistAlbumsDict.items():
        artistAlbumsByType[albumType] = getFlatList(albumsInfo.getAlbums().values())
        
    artistAlbums = []
    albumTypes = ["Random"]
    for albumType in albumTypes:
        if isinstance(artistAlbumsByType.get(albumType), list):
            artistAlbums += [artistAlbumsByType[albumType]]
    return getFlatList(artistAlbums)


def getMusicDF(mdbmap):
    mdf = mdbmap.getDF().T
    cols = mdf.columns

    mdfAlbums = DataFrame(Series({primaryKey: len(getArtistAlbums(mmb, primaryKey[0])) for primaryKey in mdbmap.getArtists()}))
    mdfAlbums.columns = ["Albums"]

    mdfSongs = DataFrame(Series({primaryKey: len(getArtistSongs(mmb, primaryKey[0])) for primaryKey in mdbmap.getArtists()}))
    mdfSongs.columns = ["Songs"]

    mdfMatches = DataFrame(Series({idx: len([x for x in row.values if x is not None]) for idx,row in mdf.iterrows()}))
    mdfMatches.columns = ["DBMatches"]

    mdfmapDF = mdf.join(mdfAlbums).join(mdfSongs).join(mdfMatches)
    mdfmapDF = mdfmapDF[["Albums"] + ["Songs"] + list(cols) + ["DBMatches"]]
    
    mdfmapDF = mdfmapDF.sort_values(by="Albums", ascending=False)
    
    return mdfmapDF
```

```python
from timeUtils import clock, elapsed
from myMusicDBMap import myMusicDBMap

class myMusicAnalysis:
    def __init__(self, mdb, debug=False):
        print("=================== Analysis of My Music ===================")
        self.debug = debug
        self.mdb   = mdb
        self.mdb.show()
        
        self.df = None
        self.results = {}

        
    def getArtistUnMatchedAlbums(self):
        retval         = {}


        ######################################################################
        #### Loop Over Prime Directories
        ######################################################################
        for primeDir in getPrimeDirectories():
            artistPrimeDirMap = getArtistPrimeDirMap(primeDir)
            print("UnMatched Albums: {0}".format(primeDir))
            #if primeDir.startswith("B"):
            #    break


            ######################################################################
            #### Loop Over Artist Name <-> Prime Map Items
            ######################################################################
            for artistName, artistPrimeDirs in artistPrimeDirMap.items():
                retval[artistName] = {"MyAlbums": [], "Todo": [], "Unknown": [], "Random": [], "Volumes": artistPrimeDirs}
                myAlbums           = {}
                for dirval in artistPrimeDirs:
                    myAlbums["MyAlbums"] = getMyMusicAlbums(dirval, returnNames=True)
                    myAlbums["Todo"]     = getMyTodoMusicAlbums(dirval)
                    myAlbums["Unknown"]  = getMyUnknownMusicAlbums(dirval)
                    myAlbums["Random"]   = getMyRandomMusic(dirval)

                    for key in myAlbums.keys():
                        retval[artistName][key] += myAlbums[key]

                for key in retval[artistName].keys():
                    retval[artistName][key] = len(retval[artistName][key])


        return retval



    def getArtistMatchedAlbums(self):
        albumTypes     = getAlbumTypes(keys=True)
        retval         = {}


        ######################################################################
        #### Loop Over Prime Directories
        ######################################################################
        for primeDir in getPrimeDirectories():
            artistPrimeDirMap = getArtistPrimeDirMap(primeDir) 
            print("Matched Albums: {0}".format(primeDir))
            #if primeDir.startswith("B"):
            #    break




            ######################################################################
            #### Loop Over Artist Name <-> Prime Map Items
            ######################################################################
            for artistName, artistPrimeDirs in artistPrimeDirMap.items():


                ######################################################################
                #### Get My Albums
                ######################################################################
                albumTypeResults = {}
                artistMatchDirs = [setDir(artistPrimeDir, 'Match', forceExist=False) for artistPrimeDir in artistPrimeDirs if isDir(artistPrimeDir, 'Match')]
                for albumType in albumTypes:
                    albumTypeResults[albumType] = []
                    for db in self.mdb.getDBs():
                        albumDBTypeNames = getAlbumTypes(db, albumType)
                        for albumDBTypeName in albumDBTypeNames:
                            for artistMatchDir in artistMatchDirs:
                                albumDBTypeDir = setDir(artistMatchDir, albumDBTypeName, forceExist=False)
                                if isDir(albumDBTypeDir):
                                    #print(albumDBTypeDir)
                                    albumTypeResults[albumType] += [getDirBasics(x)[-1] for x in findDirs(albumDBTypeDir)]
                                    #print(len(albumTypeResults),'\t',albumTypeResults)
                    albumTypeResults[albumType] = len(list(set(albumTypeResults[albumType])))
                retval[artistName] = albumTypeResults


        return retval
    
    
    def analyze(self):
        start, cmt = clock("Analyzing Artist Albums")

        artistUnMatchedAlbums = self.getArtistUnMatchedAlbums()
        artistMatchedAlbums   = self.getArtistMatchedAlbums()    

        self.results   = {}
        artistNames = set(list(artistUnMatchedAlbums.keys()) + list(artistMatchedAlbums.keys()))
        for artistName in sorted(artistNames):
            artistMatchedAlbumResults   = artistMatchedAlbums.get(artistName)
            artistUnMatchedAlbumResults = artistUnMatchedAlbums.get(artistName)
            #print(artistName,'\t',artistMatchedAlbumResults,'\t',artistUnMatchedAlbumResults)

            self.results[artistName]          = artistUnMatchedAlbumResults
            self.results[artistName]["Match"] = artistMatchedAlbumResults
            
        
        self.df = DataFrame(self.results).T            
            
        elapsed(start, cmt)
            
            
    def getDataFrame(self):
        if df is None:
            raise ValueError("DataFrame is not calculated!")
        return self.df
    
    
    def getResults(self):
        return self.results
```
